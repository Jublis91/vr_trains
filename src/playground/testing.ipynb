```python
from pathlib import Path
import duckdb
import pandas as pd
```

```python
repo_root = Path.cwd()
if not (repo_root / "data").exists():
    repo_root = repo_root.parents[1]
db_path = repo_root / "data/vr_trains.duckdb"
conn = duckdb.connect(str(db_path))
conn.execute("select * from information_schema.schemata").fetchdf()
```

```python
conn.execute("show tables").fetchdf()
```

```python
tables = conn.execute("select table_schema, table_name from information_schema.tables").fetchdf()
tables
if ((tables.table_schema == "bronze") & (tables.table_name == "live_trains")).any():
    conn.execute("select * from bronze.live_trains limit 5").fetchdf()
else:
    "bronze.live_trains not found. Run fetch_digitraffic.py and bronze_merge.py first."
tables
```
